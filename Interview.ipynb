2. Mean (numeric) and Mode (categorical), Median (numeric) and Mode (categorical)

```python
# Find numeric and categorical columns
numeric_cols = df_dirty.select_dtypes(include=[np.number]).columns
categorical_cols = df_dirty.select_dtypes(include=['object', 'category']).columns
```

```python
# Fill missing numeric values with the mean
df21 = df_dirty.copy()
df21[numeric_cols] = df21[numeric_cols].fillna(df21[numeric_cols].mean())

# Fill missing categorical values with the mode
for col in categorical_cols:
    df21[col] = df21[col].fillna(df21[col].mode()[0])

df21
```

```python
# Fill missing numeric values with the median
df22 = df_dirty.copy()
df22[numeric_cols] = df22[numeric_cols].fillna(df22[numeric_cols].median())

# Fill missing categorical values with the mode
for col in categorical_cols:
    df22[col] = df22[col].fillna(df22[col].mode()[0])

df22
```

3. KNN Imputation (numeric_categorical)

```python
# Function to impute categorical data using KNN
def knn_impute_categorical(df, column, k=5):
    # Temporarily encode missing categories as an additional label
    df_temp = df.copy()
    df_temp[column] = df_temp[column].astype('category').cat.add_categories(['missing'])
    df_temp[column].fillna('missing', inplace=True)

    # Convert categories to ordinal values
    ordinal_enc = OrdinalEncoder()
    df_temp[column] = ordinal_enc.fit_transform(df_temp[[column]]).astype(int)

    # Use KNN to impute
    knn_imputer = KNNImputer(n_neighbors=k, weights="uniform")
    df_temp[column] = knn_imputer.fit_transform(df_temp[[column]])

    # Convert back to categorical
    df_temp[column] = ordinal_enc.inverse_transform(df_temp[[column]]).astype(str)
    df_temp[column].replace('missing', np.nan, inplace=True)
    return df_temp[column]
```

```python
# Impute numeric values using KNNImputer
df_dirty_here = pd.read_csv('./sample_data/missing value/dirty_test.csv')
df3 = df_dirty_here.copy()
knn_imputer = KNNImputer(n_neighbors=5)
df3_numeric = df3.select_dtypes(include=[np.number])
df3[df3_numeric.columns] = knn_imputer.fit_transform(df3_numeric)

# Impute categorical values using KNNImputer
# df3['Cabin'] = knn_impute_categorical(df3, 'Cabin', k=5)
categorial_cols = df_dirty.select_dtypes(include=['object','category']).columns
# Impute categorical values using mode
for col in categorial_cols:
  df3[col] = df3[col].fillna(df3[col].mode()[0])

df3
```
